```python
df=sns.load_dataset("penguins")  #dataset load from github 
df
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df,hue="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df,hue="sex",size=50) #increse size
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df,hue="sex",style="sex" )
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df,hue="sex",style="sex",palette="Accent") #palette used for different color combination  
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df,hue="sex",style="sex",palette="Accent",markers=["o",">"])  
plt.show()    #markers used to point any direction 
```

```python
df=sns.load_dataset("penguins").head(20) #top 20 data
df
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df,hue="sex",style="sex",palette="Accent",markers=["o",">"])  
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df,hue="sex",style="sex",palette="Accent",markers=["o",">"],legend=False)   
plt.show() #remove lengend
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df,hue="sex",style="sex",
             palette="Accent",markers=["o",">"],legend="full")  

plt.grid()  #grid line
plt.title("python") #for title
plt.show()  #run
```
