```python
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt # type: ignore

df=sns.load_dataset("car_crashes")
n=df.drop(columns=["abbrev","ins_losses", "ins_premium"],axis=1).head(15) #drop "abbrev" column
n
```

```python
sns.heatmap(n)
plt.show()
```

```python
sns.heatmap(n,vmax=15,vmin=0, cmap="icefire") #set minimum and max value and changes color
plt.show()
```

```python
sns.heatmap(n,annot=True) #color value
plt.show()
```

```python
sns.heatmap(n,annot=True,linewidths=2) #take a width between line
plt.show()
```

```python
sns.heatmap(n,annot=True,linewidths=2,linecolor="b") #pass color between line
plt.show()
```

```python
sns.heatmap(n,annot=True,linewidths=2,cbar=False) #remove colorbar
plt.show()
```

```python
sns.heatmap(n,annot=True,linewidths=2,cbar=False,xticklabels=False,yticklabels=False) #remove x,y labels heading 
plt.show()
```

```python
v=sns.heatmap(n,annot=True,linewidths=2,cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="Seaborn",ylabel="python",title="Python library")
sns.set(font_scale=1)

plt.show()
```
