```python
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt # type: ignore

df=sns.load_dataset("penguins")#data load from github directory
df
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex")
plt.show  #basic scatter plot 
```

```python
df=sns.load_dataset("penguins").head(40)
df
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex")
plt.show() 
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",style="sex")
plt.show() #apply style 
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",size="sex")
plt.show() 
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",style="sex",sizes=(100,50))
plt.show(close=None, block=None)  #change sizes
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",style="sex",sizes=(100,50),palette="PRGn",alpha=1)
plt.show(close=None, block=None)
```

```python
mkr={"Male":"*","Female":"o"} #change markers parameter
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",style="sex",sizes=(100,50),markers=mkr)
plt.show(close=None, block=None)
```
