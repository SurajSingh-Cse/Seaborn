```python
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt

df=sns.load_dataset("penguins")#data load from github directory
df
```

```python
sns.displot(x="bill_length_mm",data=df)
plt.show() #basic syntax of histogram
```

```python
sns.displot(x="flipper_length_mm",data=df)
plt.show() 
```

```python
sns.displot(x="bill_length_mm",data=df,kde=True)
plt.show()  #used to polynomial graph
```

```python
sns.displot(x="bill_length_mm",data=df,rug=True,kde=True)
plt.show() #bottom scale 
```

```python
sns.displot(x="bill_length_mm",data=df,color="r")
plt.show() 
```
