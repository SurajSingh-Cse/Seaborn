```python
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt # type: ignore

df=sns.load_dataset("tips") #data load from github directory
df
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df,hue="sex")
plt.show()
```

```python
sns.pairplot(df,vars=["total_bill","tip"],hue="sex")
plt.show()
```

```python
sns.pairplot(df,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="BuGn")
plt.show()
```

```python
sns.pairplot(df,hue="sex",hue_order=["Female","Male"],kind="hist") #scatter,kde,hist,reg these keywords you can used in kind parameter
plt.show()
```

```python
sns.pairplot(df,vars=["total_bill","tip"],hue="sex",markers=["o","*"])
plt.show()
```
