```python
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt

df=sns.load_dataset("penguins").head(20) #data load from github directory
df
```

```python
sns.barplot(x="bill_length_mm",y="flipper_length_mm",data=df) 
plt.show() #basic bar
```

```python
sns.barplot(x="bill_length_mm",y="flipper_length_mm",data=df,hue="sex",palette="Accent")
plt.show()
```

```python
sns.barplot(x="bill_length_mm",y="flipper_length_mm",data=df,hue="sex",ci=100)
plt.show()
```

```python
sns.barplot(x="bill_length_mm",y="flipper_length_mm",data=df, orient="h")
plt.show() #flip graph horizontal,and vertical
```

```python
sns.barplot(x="bill_length_mm",y="flipper_length_mm",data=df,hue="sex",palette="OrRd")
plt.show()
```

```python
df=sns.load_dataset("penguins")
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",err_kws={'color': 'b'},palette="OrRd")
plt.show() #
```

```python
sns.set(style="darkgrid")  #background grid
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",palette="OrRd")
plt.show()
```
