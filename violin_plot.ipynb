```python
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt # type: ignore

df=sns.load_dataset("tips") #data load from github directory
df
```

```python
sns.violinplot(x="day",y="total_bill",data=df, hue="smoker",palette="Accent")
plt.show() #basic violinplot graph
```

```python
sns.violinplot(x="day",y="total_bill",data=df, hue="smoker",palette="Accent",linewidth=2)
plt.show()
```

```python
sns.violinplot(x="sex",y="total_bill",data=df, hue="smoker",palette="Accent",alpha=1,saturation=1)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=df, hue="smoker",palette="Accent",split=True)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=df,inner="quart",color="r")
plt.show()
```

```python
sns.violinplot(x=df["total_bill"]) #single plot
plt.show()
```
