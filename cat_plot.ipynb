```python
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt # type: ignore

df=sns.load_dataset("tips") #data load from github directory
df
```

```python
sns.catplot(x="size",y="tip",data=df)
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=df,hue="sex")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=df,hue="sex",kind="bar")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=df,hue="sex",kind="boxen")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=df,hue="sex",kind="point")
plt.show()
```
