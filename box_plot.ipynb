```python
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt # type: ignore

df=sns.load_dataset("tips") #data load from github directory
df
```

```python
sns.set_style("whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time")
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="time",palette='dark:g')
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="time",palette='dark:g',showmeans=True)
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="time",
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"})
plt.show()  
```
